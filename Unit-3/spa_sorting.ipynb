```python
import time
import random

# -------------------------------
# Insertion Sort
# -------------------------------
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key


# -------------------------------
# Merge Sort
# -------------------------------
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr)//2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)


def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


# -------------------------------
# Quick Sort (FIXED - Random Pivot)
# -------------------------------
def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)


def partition(arr, low, high):
    # RANDOM PIVOT (important fix)
    pivot_index = random.randint(low, high)
    arr[pivot_index], arr[high] = arr[high], arr[pivot_index]

    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i+1], arr[high] = arr[high], arr[i+1]
    return i + 1


# -------------------------------
# Timing Function
# -------------------------------
def measure_time(func, arr):
    temp = arr.copy()

    start = time.time()

    if func == merge_sort:
        func(temp)
    elif func == quick_sort:
        func(temp, 0, len(temp) - 1)
    else:
        func(temp)

    end = time.time()
    return round((end - start) * 1000, 2)  # ms


# -------------------------------
# Dataset Generator
# -------------------------------
def generate_data(n):
    random_list = [random.randint(1, 100000) for _ in range(n)]
    sorted_list = sorted(random_list)
    reverse_list = sorted_list[::-1]

    return random_list, sorted_list, reverse_list


# -------------------------------
# Correctness Check
# -------------------------------
def correctness_check():
    test = [5, 2, 9, 1, 5, 6]

    print("Correctness Check:")
    print("Input :", test)

    a = test.copy()
    insertion_sort(a)
    print("Insertion Sort:", a)

    b = merge_sort(test.copy())
    print("Merge Sort    :", b)

    c = test.copy()
    quick_sort(c, 0, len(c) - 1)
    print("Quick Sort    :", c)


# -------------------------------
# Main Execution
# -------------------------------
def main():
    correctness_check()

    sizes = [1000, 5000, 10000]

    print("\n--- PERFORMANCE RESULTS ---")

    for size in sizes:
        rand, sorted_arr, rev = generate_data(size)

        print(f"\nSize: {size}")

        # Random
        print("Random:",
              measure_time(insertion_sort, rand),
              measure_time(merge_sort, rand),
              measure_time(quick_sort, rand))

        # Sorted
        print("Sorted:",
              measure_time(insertion_sort, sorted_arr),
              measure_time(merge_sort, sorted_arr),
              measure_time(quick_sort, sorted_arr))

        # Reverse
        print("Reverse:",
              measure_time(insertion_sort, rev),
              measure_time(merge_sort, rev),
              measure_time(quick_sort, rev))


# -------------------------------
# Run
# -------------------------------
if __name__ == "__main__":
    main()
```
